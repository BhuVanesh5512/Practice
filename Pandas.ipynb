```python
import pandas as pd
```

```python
#Series :1-D datatype of labelled array (column) in a spreadsheet
data1 = [100,102, 104]#int
data2 = ["A","B","c"]#string =object
data3 =[True,False]#boolean
series = pd.Series(data1)
series1 =pd.Series(data2)
series2 =pd.Series(data3)
print(series)
print(series1)
print(series2)
```

```python
series
```

```python
series =pd.Series(data1, index =["a","b","c"])#index change / label change
print(series)
```

```python
print(series.loc["a"])#used to retrive from the value by label (location)
series.loc["c"]= 200
print(series)
```

```python
print(series.iloc[0])#used by integer label
```

```python
data = [100,200,300,400,500]#filtering by value
series3 = pd.Series(data , index = ["a","b","c","d","e"])
print(series3[series3 > 200])
```

```python
data4 = {"Day1": 1750, "Day2": 2100, "Day3":1700}
series4 = pd.Series(data4)
print(series4.loc["Day1"])
print(series4.iloc[2])
series4.loc["Day1"] +=400
print(series4)
print(series4[series4 >2000])
```

```python
#Datarframe = Atabular data Structure with rows and columns.(2-D) same as spread sheet
data = {"Name" : ["Luffy", "Zoro", "Sanji"],
        "Age" : [19 ,22, 22]}
df = pd.DataFrame(data, index =["Captian", "Vice-Captian", "Cook"])
print(df)
```

```python
print(df.loc["Captian"])#For Single Row by label
```

```python
print(df.iloc[1])#same with index
print(df.iloc[2])
```

```python
#Add new Column
df["Joining"] = ["Captian", "Crew1", "Crew3"]
print(df)
```

```python
#Add new Row : create new dataframe and concatenate
new_row = pd.DataFrame([{"Name":"Nami", "Age": 21,"Joining" : "Crew2" }], index =["Navigator"])
df = pd.concat([df ,new_row])#used to concatenate
print(df)
```
