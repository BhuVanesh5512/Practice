```python
#using matplotlib + numpy
import numpy as np #faster to use and calculate

arrx = np.array([56,78,100,130,150])
arry = np.array([12,45,67,89,120])

plt.plot(arrx,arry)
plt.show()
```

```python
#plot customization
plt.plot(arrx,arry, marker="*")#marker is used for point customization;highlights the points for each coordinate
plt.show()#there are many markers V,O,..........
```

```python
plt.plot(arrx,arry, marker="*",markersize = 32,markerfacecolor="red",markeredgecolor = "red")#size a marker ;we can use 'ms' 
plt.show()#markerfacecolor used to customization marker color ;can use 'mfc' ;same goes to markeredgecolor "mec" used to color marker edges
```

```python
plt.plot(arrx,arry,marker= "*",mfc="red",mec="red",linestyle="dashdot")#linestyle used to customize the lines in plot
plt.show()# like dashed, dashdot,dotted...... maybe None so only markers are ploted ;default is solid
```

```python
plt.plot(arrx, arry, marker =">",mfc ='red',mec ='red',linestyle ="solid",linewidth =4,color = 'cyan')
plt.show()#linewidth are setting width and color is used to set line color
```

```python
arrz =np.array([40,23,45,61,71])
plt.plot(arrx,arry,marker ="v",mfc ='red',mec='red',linestyle ='solid',linewidth =2,color ='cyan')
plt.plot(arrx,arrz,marker =".",mfc ="blue",mec ="blue",linestyle = "solid", linewidth = 3,color ="red")
plt.show()
```

```python
#easy setup hack: if u want change color than remove key and give as argument plot
arrv=np.array([60,38,42,97,12])
line_style =dict(marker ="v",mfc ='red',mec='red',linestyle ='solid',linewidth =2,color ='cyan')
plt.plot(arrx,arry, **line_style)#** used for unpacking dictionary
plt.plot(arrx, arrz, **line_style)
plt.plot(arrx,arrv,**line_style)
plt.show()
```

```python
plt.plot(arrx, arrz)#system take different color to different plot as default
plt.plot(arrx,arrv)
plt.plot(arrx,arry)
plt.show()
```

```python
#labels & x,y customization
plt.plot(arrx,arrz)
plt.plot(arrx,arrv)
plt.plot(arrx,arry)
plt.title("Class size", fontsize =25, family ="Arial",fontweight ="bold",color = "yellow")# for title and its customization size,font,weight,color
plt.xlabel("x-coordinate",fontsize =20, family ="Arial",color ="green")
plt.ylabel("y/v/z-coordinate",fontsize =20, family ="Arial",color ="green")
plt.show()
```

```python
plt.plot(arrx,arrz)
plt.plot(arrx,arrv)
plt.plot(arrx,arry)
plt.title("Class size", fontsize =25, family ="Arial",fontweight ="bold",color = "yellow")
plt.xlabel("x-coordinate",fontsize =20, family ="Arial",color ="green")
plt.ylabel("y/v/z-coordinate",fontsize =20, family ="Arial",color ="green")
plt.tick_params(axis = "both",colors = "green")#for tick customizations axis = axis  
plt.xticks(arrx)#used to set x-coordinate as at given values
plt.show()
```

```python
#grid helps make plots erier to read by refernce lines
plt.grid()#default grid
plt.plot(arrx,arry)
plt.show()
```

```python
#grid customization
plt.grid(axis = "y", linewidth = 2,color = 'lightcyan',linestyle = "dashdot")
plt.show()
plt.grid(axis = "x", linewidth = 2,color = 'lightblue',linestyle = "dashed")
plt.show()
```

```python
#barcharts : used to compare categories as data by representing each category into bars
#making list of string don't benefits from Numpy array(additial point for numpy)
categories = np.array(["Grains","fruits","Vegetables","Dairy","Sweets"])
values = np.array([2,4,5,1,3])
plt.bar(categories,values,color ="skyblue")#bar is used to plot bar
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
categories = np.array(["Grains","fruits","Vegetables","Dairy","Sweets"])
values = np.array([2,4,5,1,3])
plt.barh(categories,values,color ="skyblue")#used for horizontal bar plot
plt.title("Daily Consumption")
plt.ylabel("Food")
plt.xlabel("Quantity")
plt.show()
```

```python
#piechart : circular chart divided into slices to show percentages of the total
categories = np.array(["Grains","fruits","Vegetables","Dairy","Sweets"])
values = np.array([2,4,5,1,3])
plt.pie(values,labels = categories )#pie is used to plot  take one argument
plt.title("Daily Consumption")
#plt.xlabel("Food")
#plt.ylabel("Quantity")
plt.show()
```

```python
categories = np.array(["Grains","fruits","Vegetables","Dairy","Sweets"])
values = np.array([2,4,5,1,3])
colors =["red","yellow","green","blue","cyan"]
plt.pie(values,labels = categories,autopct="%1.1f%%",colors=colors )#autopct is used for percentage; % used for formatting here so we use in f%%
plt.title("Daily Consumption")
#plt.xlabel("Food")
#plt.ylabel("Quantity")
plt.show()
```

```python
categories = np.array(["Grains","fruits","Vegetables","Dairy","Sweets"])
values = np.array([2,4,5,1,3])
colors =["red","yellow","green","blue","cyan"]
plt.pie(values,labels = categories,autopct="%1.1f%%",colors=colors ,explode = [0,0,0.1,0,0],shadow=True,startangle = 180)#explode used to highlight the pie slice
plt.title("Daily Consumption")#shadow used for make pie chart shadow
#plt.xlabel("Food")
#plt.ylabel("Quantity")
plt.show()#startangle rotates the pie chart in given angle
```

```python
#Scatter graph = Shows the relationship b/w two variables ;Helps to find the correlation between variables(+,-,None)
#like relation between work done and time
x1 = np.array([0, 1, 2, 3, 2, 1.5, 4, 7, 9])
y1 =np.array([55, 60, 65, 70, 65, 64, 78, 80, 100])
plt.scatter(x1,y1, color ="red")#scatter is used for ploting scatter plots
plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.title("Test Scores")
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3, 2, 1.5, 4, 7, 9])
y1 =np.array([55, 60, 65, 70, 65, 64, 78, 80, 100])
plt.scatter(x1,y1, color ="red", alpha = 0.5, s=100)# alpha used for transparency of color for dots and size 's'
plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.title("Test Scores")
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3, 2, 1.5, 4, 7, 9])
y1 =np.array([55, 60, 65, 70, 65, 64, 78, 80, 100])
x2 = np.array([0, 1, 2, 3, 2.5, 1.5, 4, 7, 8])
y2 =np.array([27, 30, 38, 55, 40, 45, 77, 89, 100])
plt.scatter(x1,y1, color ="red", alpha = 0.5, s=100,label = "class A")
plt.scatter(x2,y2, color ="cyan", alpha = 0.5, s=100,label = "class B")
plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.title("Test Scores")
plt.legend()#used to display labels of plt(class A....)
plt.show()
```

```python
#histogram:  a visual representation of data distribution
# mostly bell curves ; it make bin intervals

scores = np.random.normal(loc = 75,scale = 10,size =100)
#loc is used for location of median where  all the points cluster
#scale shows standard devation or deviation or spread of point from medium 
#no of test scores is decided by size
#random is use3d for randomaly generate values
plt.hist(scores)#hist is used for ploting histogram in matplotlib
plt.show()
```

```python
scores = np.random.normal(loc = 80,scale= 10,size =100)
# To not deviate above 100 and 0 (extremes) we use clip 
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins = 10,color = "green",edgecolor ="black")#bins invervals
plt.title("Exam Scores")
plt.xlabel("Scores")
plt.ylabel("No of students")
plt.show()
```

```python
#subplot 
"""terms: 
figure : the entrie canvas of graph
Ax : A single plot (subplot)  ; also it is numpy array """
figure, axes = plt.subplots(2, 2)
plt.show()
```

```python
#using axes to customize and plot
x = np.array([1,2,3,4,5,7,8])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x,x*2)
axes[0,0].set_title("x*2")
plt.show()
```

```python
x = np.array([1,2,3,4,5,7,8])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x,x*2)
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2,color="green")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x+12,color="red")
axes[1,0].set_title("x+12")

axes[1,1].plot(x,x**3,color="cyan")
axes[1,1].set_title("x**3")

plt.show()
```

```python
x = np.array([1,2,3,4,5,7,8])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x,x*2)
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2,color="green")
axes[0,1].set_title("x**2")

axes[1,0].bar(x,x+12,color="red")#change type of plot if u want to use
axes[1,0].set_title("x+12")

axes[1,1].plot(x,x**3,color="cyan")
axes[1,1].set_title("x**3")

plt.tight_layout()#used for dismiss overlapping
plt.show()
```
